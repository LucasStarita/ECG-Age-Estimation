```python
# ================================================
# 1. IMPORTS + CONFIGURACIÓN
# ================================================
import os
import ast
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras import mixed_precision
from google.colab import drive
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from scipy.signal import resample
import wfdb
import matplotlib.pyplot as plt

print(f"TF version: {tf.__version__}")
mixed_precision.set_global_policy("mixed_float16")
tf.config.optimizer.set_jit(True)

gpus = tf.config.list_physical_devices("GPU")
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("✔ GPU memory growth enabled")
    except Exception as e:
        print("⚠ Could not set memory growth:", e)

AUTOTUNE = tf.data.AUTOTUNE
# ================================================
# 2. MONTAR DRIVE (si lo necesitás para guardar modelos, logs, etc.)
# ================================================
drive.mount("/content/drive", force_remount=False)

# ================================================
# 3. CARGAR ARCHIVOS YA GENERADOS (orden correcto)
# ================================================
X = np.load("data_500Hz.npy")        # (21748, 5000, 12)
ages = np.load("ages.npy")           # (21748,)
meta_arr_full = np.load("meta.npy")  # (21748, 1)

print("✔ Dataset cargado")
print("X:", X.shape)
print("ages:", ages.shape)
print("meta:", meta_arr_full.shape)

# --- VERIFICACIÓN SANIDAD ---
print("Ages NaN:", np.isnan(ages).sum())
print("Meta hombres:", meta_arr_full.sum())





# ================================================
# 5. SPLIT TRAIN / VAL / TEST + ESTADÍSTICAS
# ================================================
N = X.shape[0]
idx = np.arange(N)
np.random.seed(42)
np.random.shuffle(idx)

X = X[idx]
meta_arr_full = meta_arr_full[idx]
ages = ages[idx]

train_size = int(N * 0.70)
val_size   = int(N * 0.15)
test_size  = N - train_size - val_size

X_train = X[:train_size]
X_val   = X[train_size:train_size+val_size]
X_test  = X[train_size+val_size:]

meta_train = meta_arr_full[:train_size]
meta_val   = meta_arr_full[train_size:train_size+val_size]
meta_test  = meta_arr_full[train_size+val_size:]

ages_train = ages[:train_size]
ages_val   = ages[train_size:train_size+val_size]
ages_test  = ages[train_size+val_size:]

print("Train:", X_train.shape, meta_train.shape, ages_train.shape)
print("Val  :", X_val.shape, meta_val.shape, ages_val.shape)
print("Test :", X_test.shape, meta_test.shape, ages_test.shape)

# Estadísticas de edad y meta (solo sobre train)
AGE_MEAN = float(ages_train.mean())
AGE_STD  = float(ages_train.std() + 1e-8)

meta_mean = meta_train.mean(axis=0).astype("float32")
meta_std  = meta_train.std(axis=0).astype("float32")
meta_std_safe = np.where(meta_std < 1e-6, 1.0, meta_std).astype("float32")

print(f"Edad Media (train): {AGE_MEAN:.2f}, Std: {AGE_STD:.2f}")
print("meta_mean:", meta_mean)
print("meta_std_safe:", meta_std_safe)
# ================================================
# 6. CÁLCULO DE PESOS (SAMPLE WEIGHTS) CON TRAIN
# ================================================
print("Calculando pesos manuales (train)...")
age_bins_indices = np.digitize(ages_train, bins=np.arange(0, 101, 5))
bin_counts = np.bincount(age_bins_indices, minlength=22)

max_count = np.max(bin_counts[1:])
weights_manual = np.ones(len(bin_counts), dtype=np.float32)

for i in range(1, len(bin_counts)):
    count = bin_counts[i]
    if count > 0:
        w = np.sqrt(max_count / count)
        weights_manual[i] = min(w, 5.0)  # Clamping
    else:
        weights_manual[i] = 1.0

WEIGHTS_TF = tf.constant(weights_manual, dtype=tf.float32)
print("Pesos por bin de edad:", weights_manual)
# ================================================
# 7. ARMAR tf.data DATASETS (BASE)
# ================================================
def make_base_ds(X_np, meta_np, ages_np, shuffle=True):
    ds = tf.data.Dataset.from_tensor_slices((X_np, meta_np, ages_np))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(X_np), reshuffle_each_iteration=False)
    return ds

train_base = make_base_ds(X_train, meta_train, ages_train, shuffle=True)
val_base   = make_base_ds(X_val,   meta_val,   ages_val,   shuffle=False)
test_base  = make_base_ds(X_test,  meta_test,  ages_test,  shuffle=False)
# ================================================
# 8. NORMALIZACIÓN + SAMPLE WEIGHTS + PIPELINE FINAL
# ================================================
def normalize_all(x, meta, age):
    # ECG: normalización por lead (dentro de cada muestra)
    mean = tf.reduce_mean(x, axis=0, keepdims=True)
    std  = tf.math.reduce_std(x, axis=0, keepdims=True)
    x_norm = (x - mean) / (std + 1e-6)

    # Edad target normalizada
    y = (age - AGE_MEAN) / AGE_STD

    # Meta normalizada
    meta_norm = (meta - meta_mean) / meta_std_safe
    return x_norm, meta_norm, y

def to_multimodal(x, meta, y):
    return (x, meta), y

def add_sample_weight_robust(inputs, y):
    age_real = (y * AGE_STD) + AGE_MEAN
    bin_idx = tf.cast(age_real / 5.0, tf.int32) + 1
    bin_idx = tf.clip_by_value(bin_idx, 0, len(weights_manual) - 1)
    w = tf.gather(WEIGHTS_TF, bin_idx)
    return inputs, y, w

# Aplicar normalización
train_norm = train_base.map(normalize_all, num_parallel_calls=AUTOTUNE)
val_norm   = val_base.map(normalize_all,   num_parallel_calls=AUTOTUNE)
test_norm  = test_base.map(normalize_all,  num_parallel_calls=AUTOTUNE)

BATCH_SIZE = 64

# TRAIN (pipeline infinito con weights)
train_final = train_norm.map(to_multimodal, num_parallel_calls=AUTOTUNE)
train_final = train_final.map(add_sample_weight_robust, num_parallel_calls=AUTOTUNE)
train_final = train_final.shuffle(2000)
train_final = train_final.repeat()       # repeat antes de batch
train_final = train_final.batch(BATCH_SIZE)
train_final = train_final.prefetch(AUTOTUNE)

# VAL & TEST
val_final = val_norm.map(to_multimodal, num_parallel_calls=AUTOTUNE)
val_final = val_final.batch(BATCH_SIZE).prefetch(AUTOTUNE)

test_final = test_norm.map(to_multimodal, num_parallel_calls=AUTOTUNE)
test_final = test_final.batch(BATCH_SIZE).prefetch(AUTOTUNE)

steps_per_epoch = train_size // BATCH_SIZE
print("steps_per_epoch:", steps_per_epoch)
# ================================================
# 9. MODELO: HYBRID POOLING (AVG + MAX) + META (sexo)
# ================================================
def transformer_encoder(x, num_heads, proj_dim, ff_dim, dropout=0.1):
    x_norm = layers.LayerNormalization()(x)
    attn_output = layers.MultiHeadAttention(
        num_heads=num_heads, key_dim=proj_dim, dropout=dropout
    )(x_norm, x_norm)
    x_res = layers.Add()([x, attn_output])
    x_norm2 = layers.LayerNormalization()(x_res)
    ff = layers.Dense(ff_dim, activation="gelu")(x_norm2)
    ff = layers.Dense(proj_dim, activation=None)(ff)
    x_out = layers.Add()([x_res, ff])
    return x_out

def build_model(ecg_shape=(5000, 12), meta_dim=1):

    ecg_in  = layers.Input(shape=ecg_shape, name="ecg")
    meta_in = layers.Input(shape=(meta_dim,), name="meta")

    # CNN Front-end
    x = layers.LayerNormalization()(ecg_in)
    x = layers.Conv1D(64, 7, padding="same", activation="relu")(x)
    x = layers.Conv1D(64, 7, padding="same", activation="relu")(x)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Conv1D(128, 5, padding="same", activation="relu")(x)
    x = layers.Conv1D(128, 5, padding="same", activation="relu")(x)
    x = layers.MaxPooling1D(2)(x)

    proj_dim = 128
    x = layers.Dense(proj_dim)(x)

    # Transformer
    for _ in range(3):
        x = transformer_encoder(x, num_heads=4, proj_dim=proj_dim, ff_dim=256, dropout=0.1)

    # Hybrid pooling
    avg_pool = layers.GlobalAveragePooling1D()(x)
    max_pool = layers.GlobalMaxPooling1D()(x)
    x = layers.Concatenate()([avg_pool, max_pool])

    # Meta branch
    m = layers.Dense(32, activation="relu")(meta_in)
    m = layers.Dense(16, activation="relu")(m)

    # Fusion
    h = layers.Concatenate()([x, m])
    h = layers.Dense(64, activation="relu")(h)
    h = layers.Dropout(0.3)(h)
    h = layers.Dense(32, activation="relu")(h)
    out = layers.Dense(1, dtype="float32")(h)  # salida en float32 para métricas

    return models.Model([ecg_in, meta_in], out, name="Final_Model_Weighted_PTBXL")

model_weighted = build_model()
model_weighted.summary()
# ================================================
# 10. ENTRENAMIENTO
# ================================================
optimizer_safe = tf.keras.optimizers.Adam(
    learning_rate=2e-5,
    global_clipnorm=1.0   # evita explosión de gradientes
)

model_weighted.compile(
    optimizer=optimizer_safe,
    loss="mse",
    metrics=["mae"]
)

callbacks = [
    tf.keras.callbacks.EarlyStopping(
        patience=7, monitor="val_loss", restore_best_weights=True
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss", patience=3, factor=0.3, min_lr=1e-6
    )
]

print("🟢 Entrenando model_weighted (PTB-XL)...")
history_weighted = model_weighted.fit(
    train_final,
    validation_data=val_final,
    epochs=30,
    steps_per_epoch=steps_per_epoch,
    callbacks=callbacks
)
# ================================================
# 11. EVALUACIÓN (ATÓMICA) SOBRE TEST
# ================================================
print("🔎 Extrayendo TEST en arrays estáticos...")

X_ecg_list = []
X_meta_list = []
y_list = []

for (ecg, meta), y in test_final.unbatch():
    X_ecg_list.append(ecg.numpy())
    X_meta_list.append(meta.numpy())
    y_list.append(y.numpy())

X_ecg  = np.stack(X_ecg_list, axis=0).astype("float32")
X_meta = np.stack(X_meta_list, axis=0).astype("float32")
y_true_norm = np.array(y_list).astype("float32").flatten()

print(f"✅ Datos extraídos: {X_ecg.shape}")
print("🔎 Ejecutando predicción model_weighted...")

y_pred_norm = model_weighted.predict([X_ecg, X_meta], verbose=1).flatten()

# Desnormalizar
y_pred = y_pred_norm * AGE_STD + AGE_MEAN
y_true = y_true_norm * AGE_STD + AGE_MEAN

mae  = mean_absolute_error(y_true, y_pred)
r2   = r2_score(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

print("\n🏆 RESULTADO DEFINITIVO (PTB-XL):")
print(f"✅ MAE : {mae:.4f} años")
print(f"✅ RMSE: {rmse:.4f} años")
print(f"✅ R²  : {r2:.4f}")

plt.figure(figsize=(6, 6))
plt.scatter(y_true, y_pred, alpha=0.25, s=5)
plt.plot([0, 100], [0, 100], 'r--')
plt.xlabel("Edad Real")
plt.ylabel("Edad Predicha")
plt.title(f"Modelo Ponderado + Híbrido PTB-XL (MAE: {mae:.2f})")
plt.grid(True)
plt.show()
```

2) Cargar metadata (edad + sexo + filename_hr)

3) Conversión paralela WFDB → NumPy (5000×12)

4) Guardar metadata

```python
import pandas as pd
import wfdb
import numpy as np
import os
from tqdm import tqdm
import concurrent.futures

# ===========================================================
# 1) CARGA CSV + LIMPIEZA DE EDAD NaN + META CORRECTA
# ===========================================================

df = pd.read_csv(CSV_PATH)

# 🔥 Filtrar filas con edad NaN
df = df.dropna(subset=["age"]).reset_index(drop=True)

# Edades
ages = df["age"].astype("float32").to_numpy()

# Sexo ya viene codificado: 0=female, 1=male
sex_bin = df["sex"].astype("float32").to_numpy()
meta_arr_full = sex_bin[:, None]   # shape (N,1)

# Paths WFDB completos
file_paths = df["filename_hr"].apply(lambda p: os.path.join(BASE, p)).tolist()

print("Registros:", len(df))
print("Ejemplo path:", file_paths[0])
print("Ages NaN count:", np.isnan(ages).sum())
print("Meta hombres:", meta_arr_full.sum())

# ===========================================================
# 2) CARGA PARALELA GARANTIZANDO ORDEN CORRECTO
# ===========================================================

def load_indexed(idx_path):
    i, path = idx_path
    sig, _ = wfdb.rdsamp(path)
    return i, sig.astype("float32")

N = len(file_paths)
X = np.zeros((N, 5000, 12), dtype="float32")

print("Cargando señales WFDB garantizando el orden correcto...")

with concurrent.futures.ThreadPoolExecutor(max_workers=16) as ex:
    for i, sig in tqdm(ex.map(load_indexed, enumerate(file_paths)), total=N):
        X[i] = sig

# ===========================================================
# 3) GUARDADO FINAL
# ===========================================================

np.save("data_500Hz.npy", X)
np.save("ages.npy", ages)
np.save("meta.npy", meta_arr_full)

print("✔ Señales alineadas y guardadas")
print("✔ data_500Hz shape:", X.shape)
print("✔ ages shape:", ages.shape)
print("✔ meta shape :", meta_arr_full.shape)
print("✔ Ages NaN:", np.isnan(ages).sum())
print("✔ Meta hombres (should be >0):", meta_arr_full.sum())
```
